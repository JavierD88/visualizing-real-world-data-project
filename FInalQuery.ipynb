```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies_clean
```

```python
companies_data = db.collection_companies_clean.find ({'$and': [{ 'founded_year': { '$gte': 2003 } }, { 'total_money_raised': { '$gte': 1000000.0 }}, { 'number_of_employees': { '$gte': 20 }}]})
```

```python
df = pd.DataFrame(companies_data)
```

```python
df
```

```python
df['founded_year'] = df['founded_year'].fillna(0.0).astype(int)
```

```python
geoloc_db = df[['description', 'founded_year', 'name', 'number_of_employees', 'totalOffices', 'total_money_raised', 'lat', 'lng', 'oficina_principal']]
```

```python
geoloc_db=geoloc_db.dropna()
geoloc_db.index=range(len(geoloc_db))
display(geoloc_db)
```

```python
import folium
```

```python
#mark each station as a point

m = folium.Map(location=[48.8000, 2.675], width=900, height=500, zoom_start=2,)

for index, row in geoloc_db.iterrows():
    #print (row['lat'], row['lng'])

    folium.CircleMarker([row['lat'], row['lng']],

                        radius=5,

                        popup="Name: {} Categoría: {}".format(row['name'],row["description"]),

                        fill_color="#3db7e4", # divvy color

                       ).add_to(m)

m.save('geo.html')
```

```python
def findof (column):
    lista_oficinas = []
    for e in range(len(geoloc_db)):
        lista_oficinas.append(geoloc_db.oficina_principal[e])
    return lista_oficinas

x = findof (geoloc_db.oficina_principal)
```

```python
def findNear(x, radio_max_meters):
    return list(db.collection_companies_clean.find({"oficina_principal": {"$near": {"$geometry": x, "$maxDistance": radio_max_meters,}}}))



radio_max_meters = 1000

oficinas_radio = []

for i in range(len(geoloc_db)):
    num_off = findNear(x[i], radio_max_meters)
    oficinas_radio.append(len(num_off))
    
print(oficinas_radio)

#x = findNear(geopoint, radio_max_meters)
#ist(x)
```
