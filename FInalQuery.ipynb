```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies
```

```python
companies_data = db.geoloccompanies.find ({'$and': [{ 'founded_year': { '$gte': 2003 } }, { 'total_money_raised': { '$gte': 1000000.0 } }]})
```

```python
df = pd.DataFrame(companies_data)
```

```python
df['founded_year'] = df['founded_year'].fillna(0.0).astype(int)
```

```python
geoloc_db = df[['description', 'founded_year', 'name', 'number_of_employees', 'totalOffices', 'total_money_raised', 'lat', 'lng']]
```
