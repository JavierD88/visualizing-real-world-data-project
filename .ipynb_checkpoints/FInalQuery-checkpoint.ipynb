```python
import pandas as pd
from pymongo import MongoClient
client = MongoClient('mongodb://localhost:27017/')
db = client.companies_clean
```

```python
companies_data = db.collection_companies_clean.find ({'$and': [{ 'founded_year': { '$gte': 2003 } }, { 'total_money_raised': { '$gte': 1000000.0 }}, { 'number_of_employees': { '$gte': 20 }}]})
```

```python
df = pd.DataFrame(companies_data)
```

```python
df
```

```python
df['founded_year'] = df['founded_year'].fillna(0.0).astype(int)
```

```python
df['number_of_employees'] = df['number_of_employees'].fillna(0.0).astype(int)
```

```python
geoloc_db = df[['description', 'founded_year', 'name', 'number_of_employees', 'totalOffices', 'total_money_raised', 'lat', 'lng', 'oficina_principal']]
```

```python
geoloc_db=geoloc_db.dropna()
geoloc_db.index=range(len(geoloc_db))
display(geoloc_db)
```

```python
lista = geoloc_db.oficina_principal[0]

lista
```

```python
import folium
```

```python
statmad = geoloc_db[['lat', 'lng']]
statmadlist = statmad.values.tolist()
popup = geoloc_db[['name']]
statmap = folium.Map(zoom_start=100) 

for point in range(0, len(statmadlist)):

    folium.Marker(statmadlist[point], popup=popup.iloc[point,0]).add_to(statmap)

statmap
```

```python
def findof (column):
    lista_oficinas = []
    for e in range(len(geoloc_db)):
        lista_oficinas.append(geoloc_db.oficina_principal[e])
    return lista_oficinas

x = findof (geoloc_db.oficina_principal)
```

```python
def findNear(x, radio_max_meters):
    return list(db.collection_companies_clean.find({"oficina_principal": {"$near": {"$geometry": x, "$maxDistance": radio_max_meters,}}}))



radio_max_meters = 1000

oficinas_radio = []

for i in range(len(geoloc_db)):
    num_off = findNear(x[i], radio_max_meters)
    oficinas_radio.append(len(num_off))
    
print(oficinas_radio)

#x = findNear(geopoint, radio_max_meters)
#ist(x)
```

```python
geoloc_db['offices_near'] = oficinas_radio
```

```python
geoloc_db.sort_values(by= 'offices_near', ascending=False)
```

```python
geoloc_db['Rank_employees']=geoloc_db['number_of_employees'].rank()
geoloc_db['Rank_offices_near']=geoloc_db['offices_near'].rank()
geoloc_db['Rank_money']=geoloc_db['total_money_raised'].rank()

geoloc_db['top_offices'] = geoloc_db['Rank_employees']*0.2+geoloc_db['Rank_offices_near']*0.6+geoloc_db['Rank_money']*0.2
```

```python
import requests as req
from bs4 import BeautifulSoup
```

```python
import re
```

```python
url = 'https://www.cityfeet.com/cont/san-francisco-ca/commercial-real-estate-for-lease'
res = req.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
scrap = soup.select("address", class_ = "addressLocality")
```

```python
scrap = str(scrap)
```

```python
address = re.findall('\>.*St<', scrap)
```

```python
address = str(address)
```

```python
address = address.replace('>',"").replace('<',"")
address = address.replace('[',"").replace(']',"")
```

```python
address1 = address.split(',')
```

```python
address2 = pd.DataFrame()
address2['Address'] = address1
```

```python
address2 = address2.drop_duplicates()
```

```python
url = 'https://www.cityfeet.com/cont/san-francisco-ca/commercial-real-estate-for-lease'
res = req.get(url)
soup = BeautifulSoup(res.text, 'html.parser')
scrap = soup.select("div")
```

```python
scrap = str(scrap)
```

```python
add = re.findall('{lat.*}}', scrap)
```

```python
add = str(add)
```

```python
geolocal = add.replace('{',"").replace('}}',"")
```

```python
geolocal = geolocal.split(',')
```

```python
geolocal = str(geolocal)
```

```python
geol = re.findall('\d+.\d+|\-\d+.\d+', geolocal)
```

```python
geol = list(geol)
```

```python
latSF = []
for i in range(0,len(geol),2):
    latSF.append(geol[i])
```

```python
lngSF = []
for i in range(1,len(geol),2):
    lngSF.append(geol[i])
```

```python
localesSFrancisco = pd.DataFrame()
localesSFrancisco['lat'] = latSF
localesSFrancisco['lng'] = lngSF
```

```python
localesSFrancisco.head()
```

```python
localSF = localesSFrancisco.drop_duplicates()
```

```python
LocalesSF = pd.concat([address2, localSF], axis=1)
```

```python
LocalesSF
```

```python
statmad = LocalesSF[['lat', 'lng']]
statmadlist = statmad.values.tolist()
popup = LocalesSF[['Address']]
statmap = folium.Map(zoom_start=100) 

for point in range(0, len(statmadlist)):

    folium.Marker(statmadlist[point], popup=popup.iloc[point,0]).add_to(statmap)

statmap
```
